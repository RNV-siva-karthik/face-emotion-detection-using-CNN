```python
data=pd.read_csv('/kaggle/input/face-emotion-data/face_emotion_data.csv')
data['labels'].nunique()
```

```python
import cv2
```

```python
x=[]
images=data['images'].values
labels=data['labels'].values
for i,j in zip(images,labels):
    k=cv2.imread('/kaggle/input/face-expression-recognition-dataset/images/train/'+j+'/'+i)
    x.append(k)
len(x)==len(labels)
```

```python
Y=np.array(data['labels'])
Y=Y.astype(str)
```

```python
Y.dtype
```

```python
import matplotlib.pyplot as plt
plt.imshow(x[1])
plt.title(str(Y[1]))
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=pd.get_dummies(Y)
y=y.astype(int)
y
```

```python
x=np.array(x)/255
```

```python
y.sample()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten, Dense, Input, Dropout,BatchNormalization, Activation
from keras.regularizers import L2
```

```python
reg=L2(l2=0.02)
```

```python
model=Sequential()
model.add(Conv2D(100,(5,5),padding='same',input_shape=(48,48,3), activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))
#
model.add(Conv2D(128,(5,5),padding='same', activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))

model.add(Conv2D(256,(5,5),padding='same', activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))

'''model.add(Conv2D(256,(5,5),padding='same', activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))

model.add(Conv2D(256,(5,5),padding='same', activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))

model.add(Conv2D(256,(5,5),padding='same', activation='relu'))
model.add(MaxPooling2D((3,3),padding='same'))
model.add(Dropout(0.4))
'''

model.add(Flatten())

'''model.add(Dense(500,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(500,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))


model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))'''

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(7,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
histr=model.fit(xtrain,ytrain,epochs=100,validation_data=[xtest,ytest],batch_size=300)
```

```python
model.fit(xtrain,ytrain,epochs=100,validation_data=[xtest,ytest],batch_size=300)
```

```python
model.evaluate(xtest,ytest)
```
